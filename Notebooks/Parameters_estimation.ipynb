## Model 1: Isolated internode

```python
Gi=1.5*1e-9
def equations(vars):
    GK,GNA=vars
    eq1=El-(3*GK*E(R,T,F,Krest,Ki)+2*GNA*E(R,T,F,Nrest,Ni))/(3*GK+2*GNA)
    eq2=Gi-GK-GNA


    return [eq1,eq2]


initial_guess = [0.1,0.1]
solution = root(
    equations,
    initial_guess,
    tol=1e-16)
GK,GNA= solution.x
```

```python
print(solution.x)
```

```python
Gi-(GK+GNA)
```

```python
IA=-GK*(El-E(R,T,F,Krest,Ki))-GNA*(El-E(R,T,F,Nrest,Ni))
```

```python
print(IA)
```

```python
def equations(vars):
    Imax=vars
    eq1=IA-sum(ATPase(Imax,KmK,KmNa,Krest,Ni))

    return eq1


initial_guess = 0.1
solution = root(
    equations,
    initial_guess,
    tol=1e-16)
Imax= solution.x[0]
```

```python
print(Imax)
```

```python
def equations(vars):
    Gk,Gh,Gl=vars
    eq1=GK-Gk-Gh*(gamma_h/(1+gamma_h))-Gl*(gamma_l/(1+gamma_l))
    eq2= GNA-Gh*(1/(1+gamma_h))-Gl*(1/(1+gamma_l))
    eq3=1.875*1e-10-Gh

    return [eq1,eq2,eq3]


initial_guess = [0.1,0.1,0.1]
solution = root(
    equations,
    initial_guess,
    tol=1e-16)
Gk,Gh,Gl= solution.x
```

```python
print(solution.x)
```

```python
Gk=Gk/(P_k(R,T,F,El,Krest,Ki)*np.sqrt(Krest))
print(Gk)
```

```python
Gh=Gh/(P_h(a,b,c,El,V_half))
print(Gh)
```

```python
def equations(vars):
    GK,GNA=vars
    eq1=1.5*GK*(El-E(R,T,F,Krest,Ki))+GNA*(El-E(R,T,F,Nrest,Ni))+(Gg/4)*(El-E(R,T,F,Krest,Ki))+sum(gap_leak(El,E(R,T,F,Krest,Ki),1,Gg,Ki,Ki,Ni,Ni))
    eq2=Gi-GK-GNA


    return [eq1,eq2]


initial_guess = [0.1,0.1]
solution = root(
    equations,
    initial_guess,
    tol=1e-16)
GK,GNA= solution.x
```
